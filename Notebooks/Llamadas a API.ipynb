```python
import pandas as pd
data = pd.read_csv('../data/RickAndMortyScripts.csv')
data.head()
```

```python
def sentimentAnalysis(sentence):
    sia = SentimentIntensityAnalyzer()
    polarity = sia.polarity_scores(sentence)
    pol = polarity['compound']
    return pol
```

```python
data['sentiment_compound'] = data['line'].apply(sentimentAnalysis) ### Añado una columna que aplica a la columna 'line' la función sentimentAnalysis
data.head()
```

```python
data_ = dict(data.groupby(['name'])['sentiment_compound'].mean()) ### creo la media del sentiment_compound de cada personaje y lo paso a diccionario
data_
```

```python
df = pd.DataFrame([[key, data_[key]] for key in data_.keys()], columns=['Name', 'sentiment_compound'])
df_sorted = df.sort_values('sentiment_compound') ### paso el diccionario a df y lo ordenos por los valores del sentiment_compound
df_sorted  
```
